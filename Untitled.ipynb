```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

print("1. Loading fresh dataset...")
# Load the data fresh to get all 61,066 rows back
df = pd.read_csv('csic_database.csv')

print("2. Engineering features safely...")
# Feature 1: URL Length
df['url_length'] = df['URL'].astype(str).apply(len)

# Feature 2: HTTP Method Encoding
# Fill any missing methods with 'UNKNOWN' before encoding so it doesn't crash
df['Method'] = df['Method'].fillna('UNKNOWN')
le = LabelEncoder()
df['method_encoded'] = le.fit_transform(df['Method'])

# Feature 3: Payload Length ('lenght')
# INSTEAD OF DROPPING ROWS: Convert missing values (NaN) to 0
df['lenght'] = pd.to_numeric(df['lenght'], errors='coerce').fillna(0)

# Select final features
features = ['url_length', 'method_encoded', 'lenght']
X = df[features]

print(f"Total rows prepared: {len(X)} (Should be ~61066)\n")

print("3. Training the Isolation Forest...")
# Train the model with 5% contamination
model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
df['anomaly_score'] = model.fit_predict(X)

# Separate normal and anomalous traffic
normal_traffic = df[df['anomaly_score'] == 1]
anomalies = df[df['anomaly_score'] == -1]

print(f"Normal Traffic Detected: {len(normal_traffic)}")
print(f"Anomalies Detected: {len(anomalies)}\n")

print("4. Generating visualization...")
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(normal_traffic['url_length'], normal_traffic['lenght'], 
            c='blue', label='Normal', alpha=0.5, s=10)
plt.scatter(anomalies['url_length'], anomalies['lenght'], 
            c='red', label='Anomaly', alpha=0.8, s=20)

plt.title('API Behavior: Normal vs. Anomalous Traffic')
plt.xlabel('URL Length (Characters)')
plt.ylabel('Payload Content Length (Bytes)')
plt.legend()
plt.show()
```
